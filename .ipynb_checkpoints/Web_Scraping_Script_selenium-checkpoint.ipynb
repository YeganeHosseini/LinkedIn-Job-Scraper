- **Enter to Linkedin login page**
- **Spot the cookies pop-up and click Accept cookies**
- **Fill E-Mail Address and Password areas and click Login**
- **Click on the Jobs from the section above**
- **Search for job positions Junior Data Analyst in Germany**
- **Scroll down till the end of the page while collecting the link of each displayed Job offer**
- **Go to the next page when it is the end of the page while keep collecting links**
- **After all links are collected, go to the each link**
- **Click the “See More” button the expand the job description text.**
- **Scrape the desired data**

**Items Scraped:**

- Job Title
- Company Name
- Company Location
-  Job Description
- Work Method (Hybrid, Remote, On-Site)
- Experience Level
- Post Date

```python

import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
#Setting the PATH of Web Driver
# Driver's path
driver = webdriver.Chrome()
```

- **selenium:** Allows us to control a web browser programmatically.
- **pandas:** For storing and manipulating the extracted data.
- **time:** To pause execution, ensuring elements have time to load.

```python
# Maximize Window
driver.maximize_window() 
driver.minimize_window() 
driver.maximize_window() 
driver.switch_to.window(driver.current_window_handle)
driver.implicitly_wait(10)
```

```python
# Enter to the site
driver.get('https://www.linkedin.com/login');
time.sleep(5)
```

```python
#Accept Cookies
from selenium.webdriver.common.by import By

driver.find_element(By.XPATH, "//*[@id=\"artdeco-global-alert-container\"]/div/section/div/div[2]/button[1]").click()
```

```python
# User Credentials
# Reading txt file where we have our user credentials
with open('user_credentials.txt', 'r',encoding="utf-8") as file:
    user_credentials = file.readlines()
    user_credentials = [line.rstrip() for line in user_credentials]
user_name = user_credentials[0] # First line
password = user_credentials[1] # Second line
driver.find_element(By.XPATH, '//*[@id="username"]').send_keys(user_name)
driver.find_element(By.XPATH, '//*[@id="password"]').send_keys(password)
time.sleep(1)
```

```python
# Login button
driver.find_element(By.XPATH, '//*[@id="organic-div"]/form/div[3]/button').click()
driver.implicitly_wait(30)
```

```python
# Access to the Jobs button and click it
driver.find_element(By.XPATH, '//*[@id="global-nav"]/div/nav/ul/li[3]/a').click()
time.sleep(3)
```

```python
# Go to search results directly via link
driver.get("https://www.linkedin.com/jobs/search/?currentJobId=3975232539&keywords=Junior%20Data%20Analyst&origin=JOBS_HOME_KEYWORD_AUTOCOMPLETE&refresh=true")
time.sleep(1)
```

```python
try:
    links = []  # Initialize an empty list to store links
    page = 1    # Start from the first page
    
    while True:  # Loop to go through all pages
        print(f'Collecting the links on page {page}')
        
        # Find the block containing job results
        jobs_block = driver.find_element(By.CLASS_NAME, 'jobs-search-results-list')
    
        # Corrected CSS Selector to find job items within the list
        jobs_list = jobs_block.find_elements(By.CSS_SELECTOR, '.jobs-search-results__list-item')
    
        for job in jobs_list:
            # Find all <a> tags within the job element
            all_links = job.find_elements(By.TAG_NAME, "a")
        
            for a in all_links:
                href = a.get_attribute('href')
                if href and href not in links: 
                    links.append(href)  # Ensure this line is properly indented
        
            # Scroll down to the job element
            driver.execute_script("arguments[0].scrollIntoView();", job)
    
        
        print(f'Collected {len(links)} links so far.')

        # Try to find and click the 'Next' button to go to the next page
        try:
            next_page_button = driver.find_element(By.XPATH, f"//button[@aria-label='Page {page + 1}']")
            next_page_button.click()
            page += 1  # Move to the next page
            time.sleep(3)  # Wait for the next page to load
        except:
            print("No more pages left or failed to navigate to the next page.")
            break  # Exit the loop if there are no more pages

except Exception as e:
    print(f"An error occurred: {e}")

print(f"Total links collected: {len(links)}")


# Create empty lists to store information
job_titles = []
company_names = []
company_locations = []
work_methods = []
post_dates = []
work_times = [] 
job_desc = []



# Example links list, replace with your actual links
#links = ["https://example.com/job1", "https://example.com/job2"]

i = 0
j = 1
# Visit each link one by one to scrape the information
print('Visiting the links and collecting information just started.')

for i in range(len(links)):
    try:
        driver.get(links[i])
        time.sleep(2)
        # Click See more.
        try:
            driver.find_element(By.XPATH, "//button[@aria-label='Click to see more description']").click()
            #driver.find_element(By.CLASS_NAME, "jobs-description__footer-button t-14 t-black--light t-bold artdeco-card__action artdeco-button artdeco-button--icon-right artdeco-button--3 artdeco-button--fluid artdeco-button--tertiary ember-view").click()
            time.sleep(2)
        except Exception as e:
            print(f"Failed to click 'See more': {e}")
        
        # Find the general information of the job offers
        contents = driver.find_elements(By.CLASS_NAME, 'p5')
        for content in contents:
            try:
                #job_titles.append(content.find_element(By.TAG_NAME, "h1").text)
                job_titles.append(content.find_element(By.CLASS_NAME, "job-details-jobs-unified-top-card__job-title").find_element(By.TAG_NAME, "h1").text)
                company_names.append(content.find_element(By.CLASS_NAME, "job-details-jobs-unified-top-card__company-name").find_element(By.TAG_NAME, "a").text)
                company_locations.append(content.find_element(By.CLASS_NAME, "job-details-jobs-unified-top-card__primary-description-container").find_elements(By.CLASS_NAME, "tvm__text")[0].text)
                #work_methods.append(content.find_element(By.CLASS_NAME, "jobs-unified-top-card__workplace-type").text)
                work_methods.append(content.find_element(By.CLASS_NAME, "ui-label").find_element(By.XPATH, ".//span[@aria-hidden='true']").text)
                post_dates.append([span.text for span in content.find_element(By.CLASS_NAME, "job-details-jobs-unified-top-card__primary-description-container").find_elements(By.CLASS_NAME, "tvm__text") if "ago" in span.text][0])
                work_times.append(content.find_element(By.CLASS_NAME, "job-details-jobs-unified-top-card__job-insight-view-model-secondary").find_element(By.XPATH, ".//span[@aria-hidden='true']").text)
                print(f'Scraping the Job Offer {j} DONE.')
                j += 1
            except Exception as e:
                print(f"Failed to scrape general information: {e}")
            time.sleep(2)
        
        # Scraping the job description
        job_description = driver.find_elements(By.CLASS_NAME, 'jobs-description__content')
        for description in job_description:
            try:
                job_text = description.find_element(By.CLASS_NAME, "jobs-box__html-content").text
                job_desc.append(job_text)
                print(f'Scraping the Job Offer {j}')
                time.sleep(2)
                j += 1
            except Exception as e:
                print(f"Failed to scrape job description: {e}")
    except Exception as e:
        print(f"Failed to visit the link {links[i]}: {e}")

# Creating the dataframe 
df = pd.DataFrame(list(zip(job_titles, company_names,
                    company_locations, work_methods,
                    post_dates, work_times)),
                    columns=['job_title', 'company_name',
                             'company_location', 'work_method',
                             'post_date', 'work_time'])

# Storing the data to CSV file
df.to_csv('job_offers.csv', index=False)

# Output job descriptions to txt file
with open('job_descriptions.txt', 'w', encoding="utf-8") as f:
    for line in job_desc:
        f.write(line)
        f.write('\n')

print("Scraping completed and data saved.")
```

- **webdriver.Chrome():** Initializes a Chrome browser controlled by Selenium.
- **driver.get():** Opens the specified LinkedIn job search page.
- **time.sleep():** Pauses the script for a few seconds to allow the page to load fully.

- **find_elements():** Retrieves all job cards on the page.
- **job.click():** Clicks on each job card to load detailed job information.
- **driver.find_element():** Extracts specific data like the job title, company, location, and job description.
- **driver.quit():** Closes the browser after scraping is complete.
